```python
import statistics as st
import random as rd
import pandas as pd
import plotly.express as px
import random as rd
import plotly.figure_factory as ff
```

```python
df = pd.read_csv("/content/medium_data.csv")
data = df['claps'].tolist()
data.pop(0)

fig = ff.create_distplot([data],["Claps"],show_hist=False)
fig.show()

mean = st.mean(data)
standard_deviation = st.stdev(data)

print("")
print("{} is Population Mean".format(mean))
print("{} is Population Standard Deviation".format(standard_deviation))
```

```python
df = pd.read_csv("/content/medium_data.csv")
data = df['claps'].tolist()
data.pop(0)

samplingMean = st.mean(data)
sampleingstdev = st.stdev(data)

def random_set_of_mean(counter):
  dataset=[]
  for i in range (0,100):
    random_index = rd.randint(0,len(data)-1)
    value = data[random_index]
    dataset.append(value)

  meanofsample =  st.mean(dataset)
  return meanofsample


def setup():
  mean_list=[]
  for i in range(0,2578):
    set_of_means=random_set_of_mean(100)
    mean_list.append(set_of_means)
  samplingMean = st.mean(mean_list)
  sampleingstdev = st.stdev(mean_list)

  fig=ff.create_distplot([mean_list],["temp"],show_hist=False)
  fig.show()  

  print("")
  print("mean of sampling distribution: ",samplingMean)
  print("standard deviation of sampling distribution: ",sampleingstdev)

setup()
```
